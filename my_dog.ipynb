```python
import tensorflow as tf
from tensorflow import keras
import cv2
import matplotlib.pyplot as plt
```

```python
meepo =cv2.imread("meepo13.JPG")
```

```python
plt.imshow(meepo)
```

```python
meepo.shape
```

```python
meepo = meepo.reshape(1, 1472, 828, 3)/255.0
meepo.shape
```

```python
meepo_new = tf.image.resize(meepo, [224, 224])
```

```python
dense = keras.applications.DenseNet201()
```

```python
y_prob = dense.predict(meepo_new)
```

```python
keras.applications.densenet.decode_predictions(y_prob, top=5)
```
